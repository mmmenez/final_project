## 2018_BGG_cleaner_code

```python
# Imports
import pandas as pd
```

```python
# I had to add in the engine argument to properly read the files
df = pd.read_csv('2018_BGG/bgg_db_1806.csv', engine = 'python')
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
# Converting all strings to lowercase for easier comparison with the KS dataset later on
df = df.applymap(lambda s:s.lower() if type(s) == str else s)
```

```python
# Dummifying the game mechanics column
# Note that we use str.get_dummies to break apart the strings of mechanics in the column,
# which does not have the drop_first = True argument
mechdum2018 = df["mechanic"].str.get_dummies(", ")
```

```python
mechdum2018.head(3)
```

```python
# Merging the dummified mechanics into the dataframe
df = pd.concat([mechdum2018, df], axis=1, ignore_index=False, sort=True)
```

```python
# Dummifying the game categories column
# Note that we use str.get_dummies to break apart the strings of mechanics in the column,
# which does not have the drop_first = True argument
catdum2018 = df["category"].str.get_dummies(", ")
```

```python
# Merging the dummified categories into the dataframe with the dummified mechanics
df = pd.concat([catdum2018, df], axis=1, ignore_index=False, sort=True)
```

```python
df.head()
```

```python
# changing the 'names' column into 'name' to merge on with the KS data later on
df['name'] = df ['names']
```

```python
# Saving the cleaned up BGG data as a CSV for the model
df.to_csv("BGG_Data/2018_06_BGG_cleaned.csv", index = False)
```
